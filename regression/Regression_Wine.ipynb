```python
from sklearn import linear_model
import matplotlib.pyplot as plt
import matplotlib as mpl
import numpy as np
import pandas

train = pandas.read_csv("wine.csv")
test = pandas.read_csv("wine_test.csv")
```

```python
mpl.rcParams['figure.subplot.hspace'] = 0.3
mpl.rcParams['figure.figsize'] = 16,8

fig, axes = plt.subplots(nrows=3, ncols=2, sharey=True)

axes[0,0].set_title('WinterRain')
axes[0,0].scatter(train["WinterRain"], train["Price"],color='black')
axes[0,1].set_title('AGST')
axes[0,1].scatter(train["AGST"], train["Price"],color='black')
axes[1,0].set_title('HarvestRain')
axes[1,0].scatter(train["HarvestRain"], train["Price"],color='black')
axes[1,1].set_title('Age')
axes[1,1].scatter(train["Age"], train["Price"],color='black')
axes[2,0].set_title('FrancePop')
axes[2,0].scatter(train["FrancePop"], train["Price"],color='black')

plt.show()
```

```python
mpl.rcParams['figure.figsize'] = 8,4

# Create linear regression object
regr = linear_model.LinearRegression()
single_predictor = "AGST"
train_data = train.loc[:,[single_predictor]]
regr.fit(train_data, train["Price"])

# The coefficients
print('Coefficients: \n', regr.coef_)
# The mean square error
print("Residual sum of squares: %.2f"
      % np.mean((regr.predict(train_data) - train["Price"]) ** 2))
# Explained variance score: 1 is perfect prediction
print('Variance score: %.2f' % regr.score(train_data, train["Price"]))

# Plot outputs
plt.scatter(train[single_predictor], train["Price"],  color='black')
plt.plot(train[single_predictor], regr.predict(train.loc[:,[single_predictor]]), color='blue',
         linewidth=3)
plt.show()
```
